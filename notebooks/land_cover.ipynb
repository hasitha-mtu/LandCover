```python
import os
import rioxarray as rxr
from rioxarray.merge import merge_arrays
import matplotlib.pyplot as plt
```

```python
cwd = os.getcwd()
cwd
```

```python
import glob
import rasterio
```

```python
images = glob.glob("../plot/*.jp2")
images
```

```python
with rasterio.open('../plot\\B12_20m.jp2') as b12:
        b12_data = b12.read(1)
with rasterio.open('../plot\\Average.jp2') as average:
        average_data = average.read(1)
with rasterio.open('../plot\\BiLinear.jp2') as bl:
        bl_data = bl.read(1)
with rasterio.open('../plot\\Cubic.jp2') as cubic:
        cubic_data = cubic.read(1)
with rasterio.open('../plot\\Gauss.jp2') as gauss:
        gauss_data = gauss.read(1)
with rasterio.open('../plot\\NN.jp2') as nn:
        nn_data = nn.read(1)
```

```python
gauss_data
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(10, 10))

axes[0, 0].imshow(b12_data, cmap='viridis')
axes[0, 0].set_title('B12_20m')

axes[0, 1].imshow(average_data, cmap='viridis')
axes[0, 1].set_title('Average')

axes[1, 0].imshow(bl_data, cmap='viridis')
axes[1, 0].set_title('BiLinear')

axes[1, 1].imshow(nn_data, cmap='viridis')
axes[1, 1].set_title('Nearest Neighbour')

axes[2, 0].imshow(gauss_data, cmap='viridis')
axes[2, 0].set_title('Gauss')

axes[2, 1].imshow(cubic_data, cmap='viridis')
axes[2, 1].set_title('viridis')

plt.tight_layout(pad=1.00)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 10))

axes[0].imshow(b12_data, cmap='viridis')
axes[0].set_title('B12_20m')

axes[1].imshow(average_data, cmap='viridis')
axes[1].set_title('Average')
plt.tight_layout()
plt.show()
```
